Replaces path OpenDSS looks for load shapes when first running, it gets replaced with random allocation. The file linked is never used but is needed for OpenDSS to run.

```python
def OpenFile(Net,Feed):
    try:
        f = open(f"NandoLV/LVNetworkModels/Network_{Net}/Feeder_{Feed}/LoadShapes.txt", 'r')
        return True, f
    except:
        return False, 0
```

```python
# Update loadShapes.txt file
# Replace with file path
txt = "/Users/OMalleyAG/Documents/Python"

Networks = range(1,27)

for Net in Networks:
    Feed = 0
    while True:
        Feed += 1
        print(Net, Feed)
        # exception on opening
        Excepted, f = OpenFile(Net,Feed)
         
        if not Excepted:
            break

        T = f.read()
        #replace address with load file location
        lines = len(T.split("\n"))
        Ans = []
        for i in range(1,lines+1):
            Ans.append(f"New Loadshape.Shape_{i} npts=288 interval=0.0833 mult=(file={txt}/NandoLV/Summer_Load_CSVs/Customer_1.csv) useactual=false")

        X = "\n".join(Ans)
        f.close()

        with open(f"NandoLV/LVNetworkModels/" +
        f"Network_{Net}/Feeder_{Feed}/LoadShapes.txt", "w") as file:
            file.write(X)
```
